a) Grafica la llamada curva deltoide, definida parametricamente por las ecuaciones:

$x = 2 cos  + cos 2,\hspace{1 cm} y = 2 sin   sin 2$

donde 0   < 2. Toma un conjunto de valores para  entre 0 y 2 y calcula x() e y() usando las ecuaciones anteriores, para posteriormente graficar y como

```python
t = arange(0.0, 2.0*pi + 0.1, 0.1)

X = 2*cos(t) + cos(2*t)
Y = 2*sin(t) - sin(2*t)

plot(X,Y, "x-k")
xlabel("absisas")
ylabel("ordenadas")
title("Curva deltoide")
show()
```

**3. La grafica de Feigenbaum (caos determinista)**

(I) El valor se establece en un numero fijo y permanece all. Esto se llama punto fijo. Por ejemplo, $x_n = 0$ es siempre un punto fijo del mapeo logstico. (Si se pone $x_n = 0$ en el lado derecho se obtiene $x_n+1 = 0$ en el lado izquierdo).

a) Apoyate en el programa que vimos en clase y escribe un programa que muestre el comportamiento del mapeo logstico mediante una grafica.

b) De acuerdo a tu grafica, a que valor de $r$ el sistema pasa de un comportamiento ordenado (puntos fijos o ciclos lmite) a un comportamiento caotico? A este punto a veces se le llama el borde del caos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logic_map(r, x_n, it):

    # r = 5
    # x_n = 1/2

    R = [r for i in range(it+1)]
    X_n = []
    X_n.append(x_n)

    for n in range(it):
        x_n = r*x_n*(1-x_n)
        X_n.append(x_n)

    plt.plot(R,X_n,"ko", ms = 2)
```

```python
def logic_map(x_n, it):

    # r = 5
    # x_n = 1/2

    R = [i for i in np.arange(1, 4, 0.01)]

    for r in R:
      X_n = []
      X_n.append(x_n)

      RR = [r for i in range(it+1)]

      for i in range(it):
          x_n = r*x_n*(1-x_n)
          X_n.append(x_n)

      plt.plot(RR,X_n,".", ms = 1, color = "silver")
      plt.title("Feigenbaum Diagram")
      plt.xlabel("r")
      plt.ylabel("X_n")
      plt.text(1.5, 0.8, '$x_{n+1} = r x_n(1-x_n)$',horizontalalignment='center', verticalalignment='baseline')
```

```python
logic_map(0.5,1000)
```
